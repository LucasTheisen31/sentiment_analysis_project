```python
import os
import pandas as pd

from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset_name = "reviews_3000.csv"
RANDOM_SEED = 42
TEST_SIZE = 0.3
```

```python
df = pd.read_csv(dataset_name)

df.shape
df.info()
```

```python
df_train, df_test = train_test_split(df, test_size=TEST_SIZE, random_state=RANDOM_SEED)

df_val, df_test = train_test_split(df_test, test_size=0.5, random_state=RANDOM_SEED)
```

```python
print("Tamanho do conjunto de treino:", df_train.shape)
print("Tamanho do conjunto de validação:", df_val.shape)
print("Tamanho do conjunto de teste:", df_test.shape)
```

```python
# Salvar o DataFrame em um novo arquivo CSV

base_name = os.path.splitext(dataset_name)[0]

df_train.to_csv(
    f"{base_name}_dados_treino.csv",
    index=False,
    header=True,
    encoding="utf-8",
)

df_test.to_csv(
    f"{base_name}_dados_teste.csv",
    index=False,
    header=True,
    encoding="utf-8",
)

df_val.to_csv(
    f"{base_name}_dados_validacao.csv",
    index=False,
    header=True,
    encoding="utf-8",
)
```
